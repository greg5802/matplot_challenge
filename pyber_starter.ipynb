```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.patches import Ellipse

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

# Combine the data into a single dataset

# Display the data table for preview
```

```python
city_data = pd.read_csv(city_data_to_load)

ride_data = pd.read_csv(ride_data_to_load)
```

```python
ride_data_2 = ride_data.drop_duplicates(['city'])
merged_data = pd.DataFrame(pd.merge(ride_data_2,city_data, on="city"))
merged_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
merged_data2 = pd.DataFrame(pd.merge(ride_data,city_data, on="city"))
merged_data2_group = merged_data2.groupby(["city"])
merged_data2_group_df = pd.DataFrame(merged_data2_group.sum())
merged_data3 = pd.merge(merged_data2_group_df, merged_data[["city", "type"]], on="city", how='left')
merged_data3.head()

merged_data2_group_count = merged_data2_group.count()
merged_data2_gp_count_df = pd.DataFrame(merged_data2_group.count())
merged_data3_gp_count_df = pd.DataFrame(merged_data2_gp_count_df.rename(columns={'type': 'Total Rides'}))
merged_data3_gp_count_df.head()

merged_data4 = merged_data3.set_index("city")
final_merge = pd.merge(merged_data4,merged_data3_gp_count_df[["Total Rides"]], on="city", how='left')
final_merge["Average Fare $"] = round(final_merge["fare"] / final_merge["Total Rides"], 2)
final_merge.head()

urban_data = final_merge.loc[final_merge["type"] == "Urban"]
suburban_data = final_merge.loc[final_merge["type"] == "Suburban"]
rural_data = final_merge.loc[final_merge["type"] == "Rural"]
```

```python
ax = rural_data.plot(kind="scatter", x="Total Rides", y="Average Fare $", s=50, grid=True, title="Pyber Ride Sharing Data (2016)", color='gold', edgecolor='black',label="Rural")
suburban_data.plot(kind="scatter", x="Total Rides", y="Average Fare $", s=170, grid=True, title="Pyber Ride Sharing Data (2016)", ax=ax, color='skyblue', edgecolor='black', label="Suburban")
urban_data.plot(kind="scatter", x="Total Rides", y="Average Fare $", s=400, grid=True, title="Pyber Ride Sharing Data (2016)", ax=ax, color='coral', edgecolor='black', label="Urban")
```

## Total Fares by City Type

```python
# Calculate Type Percents

# Build Pie Chart

# Save Figure
```

```python
final_merge_total = final_merge["fare"].sum()
final_merge_total

urban_data_total = urban_data["fare"].sum()
urban_data_total

suburban_data_total = suburban_data["fare"].sum()
suburban_data_total


urban_per = urban_data_total / final_merge_total
suburban_per = suburban_data_total / final_merge_total
rural_per = 1 - suburban_per - urban_per

explode = (0.1,0,0)

pie = pd.DataFrame({"percentage": [urban_per,rural_per,suburban_per]}, index=["Urban","Rural","Suburban"])
plot = pie.plot.pie(y='percentage', explode = explode, figsize=(15, 5), pctdistance=0.55, colors=['coral','gold','skyblue'],autopct="%1.001f%%", shadow=True, startangle=285, legend=None)
plt.axis('off')
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Build Pie Chart

# Save Figure
```

```python
final_merge_2b = final_merge["Total Rides"].sum()
urban_data_2b = final_merge.loc[final_merge["type"] == "Urban"]
urban_data_2b["Total Rides"].sum()
suburban_data_2b = final_merge.loc[final_merge["type"] == "Suburban"]
suburban_data_2b["Total Rides"].sum()
rural_data_2b = final_merge.loc[final_merge["type"] == "Rural"]
rural_data_2b["Total Rides"].sum()

urban_per_2 = (urban_data_2b["Total Rides"].sum()) / final_merge_2b
suburban_per_2 = (suburban_data_2b["Total Rides"].sum()) / final_merge_2b
rural_per_2 = (rural_data_2b["Total Rides"].sum()) / final_merge_2b

pie2 = pd.DataFrame({"percentage": [urban_per_2,rural_per_2,suburban_per_2]}, index=["Urban","Rural","Suburban",])
plot2 = pie2.plot.pie(y="percentage", explode = explode, figsize=(15, 5), pctdistance=0.55, colors=['coral','gold','skyblue'],autopct="%1.1f%%", shadow=True,  startangle=260, legend=None)
plt.axis('off')
plt.show()
```
